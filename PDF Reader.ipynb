```python
from PyPDF2 import PdfReader

PDF = PdfReader("C:\\Users\\DELL\\Downloads\\Hasna_CV_Data_Science+(2).pdf")
```

```python
# number of pages
len(PDF.pages)
```

```python
# reading a specific page from a pdf file
page_1 = PDF.pages[0]
page_1.extract_text()
```

```python
from nltk import word_tokenize

text = word_tokenize(page_1.extract_text())
```

```python
text
```

```python
from nltk.corpus import stopwords

stop_words = list(stopwords.words())
cleaned_text = [word for word in text if word.lower() not in stop_words]
cleaned_text = [word.lower() for word in cleaned_text if word.isalpha()]
```
