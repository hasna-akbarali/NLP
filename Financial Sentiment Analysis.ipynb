```python
import pandas as pd
import numpy as np
from nltk.stem import WordNetLemmatizer
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:\\Users\\DELL\\Downloads\\financial_news_sentiment.csv',encoding='ISO-8859-1',names=['sentiment','text'])
df
```

```python
df.sentiment.value_counts(normalize=True)
```

```python
df['sentiment'] = df['sentiment'].map({'neutral':0,'positive':1,'negative':-1})
```

```python
df.sentiment.value_counts(normalize=True)
```

```python
X = df['text']
y = df['sentiment']
```

```python
import string
import re
# preprocessing our text

# lower the case
X = [i.lower() for i in X]

# remove punctuations
X=[''.join(c for c in x if c not in string.punctuation) for x in X]# remove punctuation marks

# remove numbers
X=[''.join(c for c in x if c not in '0123456789') for x in X]# remove punctuation marks

# remove whitespace
X = [i.strip() for i in X]
X
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer

cv = CountVectorizer(stop_words='english')
X_cv = cv.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X_cv,y,random_state=12,test_size=0.2,stratify=y)
```

```python
import nltk

def tokenizer(text):
    words = nltk.tokenize(text)
    return words
```

```python
from sklearn.naive_bayes import MultinomialNB

nb = MultinomialNB()
```

```python
nb.fit(X_train,y_train)
```

```python
y_pred = nb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_pred,y_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
def enter_text(msg):
    if nb.predict(cv.transform([msg])) == 0 :
               print('Neutral')
    elif nb.predict(cv.transform([msg])) == 1 :
                 print('Positive')
    else:
                 print('Negative')
```
